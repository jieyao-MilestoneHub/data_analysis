```python
# 讀取 .dat
import pandas as pd

df = pd.read_table("C:\\Users\\jiao\\Downloads\\spbnd.dat", sep = '::', header=None, engine='python')
```

```python
# 找出幾點
val = 1.01950568
xyz_info={}
size=5

for i in range(22):
    for j in range(232):
        
        data_now = df.loc[i*232+j][0].split()
        
        if float(data_now[1]) < val:
            pass
        else:
            if float(data_now[1]) > val:
                xyz_info[data_now[0]] = {'x':float(data_now[5]),'y':float(data_now[8]),'z':float(data_now[11])}
                break
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_comp_fi(xyz_info, x_=0, size=1, points=21):  # xyz_info : xyz information; x_ : 要畫在 (x,i,0) for 1 <= i <= points ; size : the size of boudnary
    
    fig1 = plt.figure(figsize=(15,28))
    ax1 = fig1.add_subplot(111, projection='3d')

    X,Y = np.meshgrid(np.linspace(-1*size,size,10), np.linspace(-1*size,size,10))

    Z = X*0

    # 繪製平面
    ax1.plot_surface(X, Y, Z, alpha=0.01)
    ax1.set_zlim(-1*size,size)
    
    
    disc_pt = np.linspace(-1*size,size,points)
    ite = 0
    for i in xyz_info.values():
        ax1.scatter(x_, disc_pt[ite], 0, color='r', s=20)
        u = i['x']
        v = i['y']
        w = i['z']
        ax1.quiver(x_, disc_pt[ite], 0, u, v, w, color='k', length=0.1)
        ite+=1

    # 隐藏坐标轴
    ax1.set_axis_off()

    # 設定邊界
    ax1.plot([-1*size,size],[size,size],[0,0], 'k',  alpha=0.2)
    ax1.plot([-1*size,-1*size],[-1*size,size],[0,0], 'k', alpha=0.2)
    ax1.plot([-1*size,size],[-1*size,-1*size],[0,0], 'k',  alpha=0.2)
    ax1.plot([size,size],[-1*size,size],[0,0], 'k', alpha=0.2)

    # 調整視角
    ax1.view_init(elev=-15, azim=45)

    ax1.set_xlabel('X')
    ax1.set_ylabel('Y')
    
    plt.show()
```

```python
x_ = np.linspace(-size,size,20)
plot_comp_fi(xyz_info, x_=x_)
```
